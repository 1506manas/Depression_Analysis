```python
import pandas as pd
import numpy as np
```

```python
import re
import nltk
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
from nltk.corpus import stopwords
import string
stopword = stopwords.words("english")
```

```python
df = pd.read_csv("Data/Mental-Health-Twitter.csv")
df.head()
```

```python
df = df[['post_text', 'label']]
```

```python
df
```

```python
df['label'].unique()
```

```python
df['label'].replace((0,1), ['Sentence found to be Depressed', 'Not Depressed'], inplace = True)
df
```

```python
def clean(text):
    text = text.lower()
    text = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?|(\n)|(\w*\d\w*)", "", text)
    text = re.sub(r'[^\w\s]', '', text)
    text = [word for word in text.split(' ') if word not in (stopword)]
    text = " ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text = " ".join(text)
    return text
df['post_text'] = df['post_text'].apply(clean)
df.head()
```

```python
import csv
from translate import Translator

def translate_csv(input_file, output_file):
    with open(input_file, 'r') as f:
            reader = csv.reader(f)
            data = list(reader)
        
    translator = Translator(to_lang="hi")

    translated_data = []
    for row in data:
        translated_row = []
        for cell in row:
            try:
                translated_cell = translator.translate(cell)
                translated_row.append(translated_cell)
            except StopIteration:
                print(f"Error translating text: {cell}")
        translated_data.append(translated_row)

    with open(output_file, 'w') as f1:
        writer = csv.writer(f1)
        writer.writerows(translated_data)

input_file = "Data/final.csv"     #input_file = df
output_file = "/content/drive/MyDrive/depression/dep_hindi1.csv"
translate_csv(input_file, output_file)
```
